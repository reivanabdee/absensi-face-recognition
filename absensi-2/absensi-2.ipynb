```python
import cv2
import numpy as np
from playsound import playsound

net = cv2.dnn.readNet("dnn_model/yolov4-tiny.weights", "dnn_model/yolov4-tiny.cfg")
model = cv2.dnn_DetectionModel(net)
model.setInputParams(size =(320, 320), scale = 1/255)
video = cv2.VideoCapture('https://192.168.100.154:8080/video')
# video = cv2.VideoCapture(0)
# video.set(cv2.CAP_PROP_FRAME_WIDTH, 720)
# video.set(cv2.CAP_PROP_FRAME_HEIGHT, 640)

classes = []
with open("dnn_model/classes.txt", "r") as file_oblect:
    for class_name in file_oblect.readlines():
        class_name = class_name.strip()
        classes.append(class_name)
                        ##fungsi mouse klik        
# def mouse_click(event,x ,y ,flags, params):
#     if event == cv2.EVENT_LBUTTONDOWN:
#         print (x,y)
# cv2.namedWindow("frame")
# cv2.setMouseCallback("frame", mouse_click)

# area = np.array([[380,0], [380,479]] )
                        ##titik tengah
# def center_handle(x,y,w,h):
# 	x1= int(w/2)
# 	y1= int(h/2)
# 	cx= x+x1
# 	cy= y+y1
# 	return cx,cy
# detec = []

## YOLO v4
warn = []
while True:
    ret, frame = video.read()
    frame = cv2.resize(frame, (640, 480), interpolation= cv2.INTER_LINEAR)
    roi = frame[121: 387,212: 438]
    area = [(212,121),(212,387),(438,387),(438,121)]
    # peringatan = clasnam
    ##area = [(x,y)kiri atas,(x,y)kiri bawah,(x,y)kanan bawah,(x,y)kanan atas]    
    
    for areas in [area]:
        cv2.polylines(frame, [np.array(areas, np.int32)], True, (255,0,0), 5)
        
        (class_ids, scores, bboxes) = model.detect(roi)
        for class_id, score, bbox in zip(class_ids, scores, bboxes):
            (x , y, w, h) = bbox
            class_name = classes[class_id] 
            cx = int(x + (w/2))
            cy = int(y + (h/2))

    # print(score)
    # cv2.polylines(frame,[area], True, (0, 0, 255), 1)


            if class_id == 0:


                cv2.putText(roi, class_name , (x, y -10), cv2.FONT_HERSHEY_PLAIN, 1,(0,0,0),2)
                cv2.rectangle(roi, (x,y), (x+w,y+h), (0,255,0), 3)
                cv2.circle(roi, (cx,cy) ,5 , (0,255,255),2)

                if score >=0.9:
                    print("terdeteksi")


            elif class_id == 67:

                cv2.putText(roi, class_name , (x, y -10), cv2.FONT_HERSHEY_PLAIN, 1,(0,0,0),2)
                cv2.rectangle(roi, (x,y), (x+w,y+h), (0,0,255), 3)
                cv2.circle(roi, (cx,cy) ,5 , (0,255,255),2)
           

    cv2.imshow("frame", frame)
    # cv2.imshow("roi", roi)
    
    if cv2.waitKey(1) & 0xFF == ord ('q'):
        break
        
video.release()
cv2.destroyAllWindows()        
```
